```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt

# Load NDVI GeoTIFF
file_path = "../data/CA_VEG_Sentinel2_NDVI.tif"

with rasterio.open(file_path) as src:
    ndvi = src.read(1)  # Read the NDVI band


# Normalize NDVI to range [0, 1]
Pveg = (ndvi - np.min(ndvi)) / (np.max(ndvi) - np.min(ndvi))

# Display Vegetation Factor (Pveg)
plt.imshow(Pveg, cmap="YlGn")
plt.colorbar(label="Vegetation Factor (Pveg)")
plt.title("Vegetation Factor Map (NDVI-based Pveg)")
plt.show()
```
